# check for null values

```python
data.isnull().sum()
```

# Filling the missing values

```python
data['Age'] = data['Age'].fillna(data['Age'].median())

data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

print(data.isnull().sum())
```

# visualizations

```python
plt.figure(figsize=(12, 5))
sns.histplot(data["Age"], kde=True, bins=30, color="#C085D4")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(data["Fare"], kde=True, bins=30, color="#C085D4")
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(data=data, x="Parch", color="#C085D4")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=data, x='Survived', palette=["#6B3F69", "#C085D4"])
plt.title("Survived vs Not Survived")
plt.show()
```

```python
sns.countplot(data=data, x="Sex", hue="Survived",
              palette={0: "#6B3F69", 1: "#C085D4"})

plt.title("Survival by Sex")
plt.show()
```

```python
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})

sns.barplot(x="Pclass", y="Fare", hue="Sex", data=data, palette={0: "#6B3F69", 1: "#C085D4"})

plt.title("Average Fare by Class and Gender")
plt.xlabel("Passenger Class (1st, 2nd, 3rd)")
plt.ylabel("Average Fare Paid")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
