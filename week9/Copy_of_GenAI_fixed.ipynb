**📝 Reflection:**

Which tokenizer produced more tokens?
**answer** **BERT** produced more tokens than GPT-2 for the same text.


Why?
**answer** BERT uses **WordPiece** — it splits rare words into sub-pieces (e.g., `transforming` → `transform` + `##ing`).
GPT-2 uses **BPE** — it keeps common words whole, so it produces fewer, larger tokens.

## Section 3 — RAG Fundamentals (Day 4)
**What is Retrieval Augmented Generation (RAG)?**
-  Retrieval + Generation = AI answers using your own data.

**answer:** The RAG **retrieves** relevant documents from data, then **generates** an answer using that context.
Without RAG, LLMs only know what they were trained on. With RAG, they can answer from documents.


**Why LLMs forget information**

**answer:** LLMs have a fixed training they don't know anything added after that date, and they have no access to private or custom data. RAG solves this by injecting fresh context at query time.

**Embeddings concept**

A **vector** (list of numbers) that represents the *meaning* of a piece of text. Similar sentences produce similar vectors.

**Vector search**
Finds the document whose embedding is *closest* to the query embedding — meaning the most semantically relevant result.


```python
# Step 1: Create Documents
docs = [
    "AI stands for Artificial Intelligence.",
    "Machine learning is a subset of AI.",
    "GANs generate synthetic data."
]
```

```python
# Step 2: Create Embeddings (FREE)
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer('all-MiniLM-L6-v2')

doc_embeddings = embed_model.encode(docs)
```

```python
# Step 3: Search Relevant Text
import numpy as np

query = "What is AI?"
query_embedding = embed_model.encode([query])

scores = np.dot(doc_embeddings, query_embedding.T)

best_doc = docs[np.argmax(scores)]
print(best_doc)
```

## Section 5 — GAN Fundamentals (Day 6)


**Topics**

**What are GANs?**

two neural networks compete to generate realistic synthetic data.

**Generator vs Discriminator**
| | Generator | Discriminator |
|---|---|---|
| **Job** | Create fake data | Detect fake vs real |
| **Input** | Random noise | Real or fake sample |
| **Output** | Fake image/data | Probability (real or fake) |
| **Goal** | Fool the Discriminator | Not be fooled |

**Why GANs mattered historically**
 GANs were the first method to generate *photorealistic* images from noise. They revolutionized image synthesis before diffusion models took over.

**GAN vs Diffusion comparison
 (mention work of Ian Goodfellow)**
| | GAN | Diffusion |
|---|---|---|
| **Quality** | Good | State-of-the-art |
| **Training** | Unstable | Stable |
| **Speed** | Very fast | Slow |
| **Used today?** | ⚠️ Declining | ✅ Dominant |

**Why GANs are harder**
Training two networks simultaneously is unstable. The Generator can hit **mode collapse** — producing the same output repeatedly instead of diverse results.

**When GANs are still useful**
- Fast data augmentation
- Lightweight synthetic data generation
- Style transfer (CycleGAN)
- When inference speed matters more than quality

### Simple GAN (PyTorch)


```python
import torch
import torch.nn as nn
```

#### Generator


```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(100,128),
            nn.ReLU(),
            nn.Linear(128,784),
            nn.Tanh()
        )

    def forward(self,x):
        return self.model(x)
```

#### Discriminator



```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(784,128),
            nn.ReLU(),
            nn.Linear(128,1),
            nn.Sigmoid()
        )

    def forward(self,x):
        return self.model(x)
```

**📝 Task:**

Explain generator vs discriminator roles.

The **Generator** takes random noise as input and outputs a fake image. Its goal is to produce outputs realistic enough to fool the Discriminator.

The **Discriminator** takes an image (real or fake) and outputs a score — how likely it is to be real. Its goal is to correctly identify fakes.

They are trained together: the Generator gets better at faking, the Discriminator gets better at detecting, until the Generator produces near-realistic outputs.
